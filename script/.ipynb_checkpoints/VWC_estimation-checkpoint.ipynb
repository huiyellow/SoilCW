```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy import constants
import json

def read_channel_response(filename):
    '''
    Samples recorded using GNU radio file sink are stored in float32 type
    '''
    samples = np.fromfile(filename, np.complex64) # Read in file.  We have to tell it what format it is
    #samples /= 32768 # convert to -1 to +1 (optional)
    #samples = samples[::2] + 1j*samples[1::2] # convert to IQIQIQ...
    #print(samples.shape)
    return samples

def plot_signal_vector(signal_vectors):
    plt.figure(figsize=(15,10))
    color_vector = ['red','blue','green']
    color_vector_index = 0
    for aVector in signal_vectors:
        # print(np.real(aVector))
        # print(np.imag(aVector))
        plt.arrow(0,0,np.real(aVector),np.imag(aVector),lw=5,head_width=0.03, head_length=0.05,ec=color_vector[color_vector_index])
        plt.xlabel("I")
        plt.ylabel("Q")
        color_vector_index +=1

def extractEps(H_neg, H, H_pos, delta_f, d_g, cable_length, cable_speed_factor, d_m):
    tau_g = 2*d_g/constants.c
    tau_cable = 2*cable_length/(constants.c*cable_speed_factor)
    rotation_vector_neg = np.exp(-2j*np.pi*delta_f*(tau_g+tau_cable))
    roated_H_neg = H_neg*rotation_vector_neg
    rotation_vector_pos = np.exp(2j*np.pi*delta_f*(tau_g+tau_cable))
    roated_H_pos = H_pos*rotation_vector_pos

    H_d_1 = roated_H_neg - H
    H_d_2 = roated_H_pos - H

    H_ratio = H_d_1/H_d_2
    phi = np.angle(H_ratio)+np.pi
    #print(phi)

    eps_a = (constants.c*phi)/(4*np.pi*delta_f*d_m)
    #print(eps_a*eps_a)
    return eps_a*eps_a

def topp_equation(dielectric_constant):
    """
    Calculates volumetric soil water content using the Topp equation.
    
    Args:
    dielectric_constant: The relative dielectric constant of the soil.
                         This is a measure of how well the soil conducts
                         electricity. It's usually measured with a special
                         sensor.
    
    Returns:
    The volumetric soil water content (θ) as a decimal (between 0 and 1).
    Returns None if the dielectric constant is negative (which shouldn't happen
    in reality).
    """
    if dielectric_constant < 0:
        return None
    theta = (4.3 * 10**-6) * (dielectric_constant**3) - \
          (5.5 * 10**-4) * (dielectric_constant**2) + \
          (2.92 * 10**-2) * dielectric_constant - 5.3 * 10**-2
    return theta

def getScanParameters(file_name):
    # We'll try to open the file in "read" mode ('r')
    with open(file_name, 'r') as f:
        # json.load() reads the data from the file and turns it back into a Python dictionary
        loaded_scan_param = json.load(f)

    # Now we can print the dictionary we loaded from the file!
    print("Loaded scan parameters from", file_name + ":")
    print(loaded_scan_param)
    return loaded_scan_param

    # # You can also access individual values from the dictionary like this:
    # main_frequency = loaded_scan_param.get("main frequency")
    # delta_f = loaded_scan_param.get("delta_f")
    # print("\nMain Frequency:", main_frequency)
    # print("Delta Frequency:", delta_f)
```

# Example of using the functions

```python
file_name = "../data/indoor_ex/indoor_potting_30cm/indoor_potting_30cm_vwc15_0.dat"
channel_responses = read_channel_response(file_name)
print("I: ", np.real(channel_responses))
print("Q: ", np.imag(channel_responses))
print("phases: ", np.angle(channel_responses))
print("amp: ", np.abs(channel_responses))
plot_signal_vector(channel_responses)

meta_file = "../data/indoor_ex/indoor_potting_30cm/indoor_potting_30cm_vwc15_meta.json"
scan_parameter = getScanParameters(meta_file)
f_c = scan_parameter.get("main frequency")
delta_f = scan_parameter.get("delta_f")
d_g = scan_parameter.get("d_g")
d_m = scan_parameter.get("d_m")
cable_length = scan_parameter.get("cable_length")
cable_speed_factor = scan_parameter.get("cable_speed_factor")

eps_a = extractEps(channel_responses[0], channel_responses[1], channel_responses[2], delta_f, d_g, cable_length, cable_speed_factor, d_m)
print(eps_a)

soil_moisture = topp_equation(eps_a)
if soil_moisture is not None:
  print(f"The aparent permittivity is: {eps_a}")
  print(f"The estimated volumetric soil water content (θ) is: {soil_moisture:.4f}")
  print(f"This means about {soil_moisture * 100:.2f}% of the soil volume is water.")
else:
  print("Invalid dielectric constant.")
```
